Q3 Difference of Static and Dynamic variables in Python ? 
Ans.  Static variables are those which are declared inside a function but outsiea ny nested function 
The Static functions are initialized at the sarting of the functiona and carry out till  the end of the program throughout the 
lifetime of the program.  

Dynamic Variables are those which are declared inside a function but within a nested function. 
So basically, the Duynamic Funtion carries out inside theneted  function and as the nested function exits the dynamic Variable 
is destroyed.  

Key Duifferences based onn different features : 
1. Scope : The Static Variables have a nested level scope whereas the dynamic functions have a neted function level scope. 
2. Initialization : The Static variables are defined when the function is initialized whereas the Dynamic variables are 
    defined when the nested function is initialized 
3. Lifetime : The Static Function is called throuhout the life time of the function of the call whereas the Dynamic Function is    called only when the nested function is called and stops as the nested function exits. 
4. Sharing: The Static Funtions are called and shared throughout the function but the dynamic function is not shared and hence it 
is only shared during the nested function execution. 



```python
# Example of the above code 
def outer_function():
    static_variable = 10 

    def inner_function():
        dynamic_variable = 20 

    inner_function() 

outer_function()
```
