Q8.  Explain the benefits of using abstract classes in object oriented design ? 

Ans. 
The benefits of using the object using abstract classses in object oriented design include : 

1. Code Reuse : Abstract classes provide a common interface and implmentation for their sub classes promoting code reuse and their
and reducing duplication. 

2. Standardization : 
Abstract classes define a standard set of methods and attributes that sub class must implement ensuring consistency and 
predictability across the code base. 

3. Polymorphism : Abstract classes enable polymorphism by allowing the subclasses to override and specialize the behavoir inherited '
 from the abstract class.  

 4. Flexibility : Abstract classes can provide default implementations for some methods, allowing the subclasses to either use 
 the default implementation or override it as needed. 

 5. Abstraction : Abstract classes help to seperate the interface from the implementation, promoting the abstracction and modularity. 

 6. Extensibility : By defining an abstract base class, you can easily add the new concrete subclasses that conform to the same 
                    interface , making code base more extensible.
