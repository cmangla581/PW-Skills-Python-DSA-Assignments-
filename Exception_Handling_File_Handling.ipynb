                                               Custom Exception 

A Custom Exception is an error class  created by developer to handle error conditions that are unique to their applications' 
logic or domain. These exceptions extend an inbuilt exception class provided by programming language.

The primary motivation for creating custom exceptions is to enhance code clarity, maintainability and provide more meaningful specific error hndling than generic inbuilt exceptions. 

1. Improved Readability : Descriptive names for custom exceptions maake the code easier for other developers to understand. 

2. Specific Error Handling : They allow for fine grained control over error handling. A catch block can specifically target a DatabaseExceptionError rather than a generic IOException.  

3. Business Logic Enforcement : Custom Exceptions are ideal for representing violations of applications specific business rules such as user trying  towithdraw more money than their account balance. 

```python
# Example with a code  

class EmptyException(RuntimeError): 
    def __init__(self, argument): 
        self.arguments = argument 

var = "" 

try : 
    raise EmptyException ("This variable is an empty string")

except EmptyException as e : 
    print(e.arguments) 
```

```python
class Num : 
    def __init__(self, num):
        self.num = num

num1 =Num(5)
num2 = Num(7) 

res = num1 + num2

# Here its showing error 
```

```python
class Num: 
    def __init__(self,num): 
        self.num = num 
    
    def __add__(self, U):
        return self.num + U.num 
    
num1 = Num(5)
num2 = Num(7) 

res = num1 + num2 

print(res)
```

```python
class ComplexNum :
 def __init__(self, x,y): 
    self.x = x
    self.y = y 

 def __add__(self,C): 
    return self.x +C.x , self.y + C.y 

c1 = ComplexNum (2,5)
c2 = ComplexNum (3,6)

print (c1 + c2) 
```

```python
class DemoComparisonOverload : 
    def __init__(self,x): 
        self.x = x 
    
    def __gt__(self,o): 
        if (self.x > o.x):
            return True 
        else : 
            return False
        
num1 =DemoComparisonOverload(4)
num2 =DemoComparisonOverload(7) 

print(num1 > num2)
```
