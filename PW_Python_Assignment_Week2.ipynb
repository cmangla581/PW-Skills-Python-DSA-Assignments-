Q What are operators / Explain all the 7 operators in details with examples.
Ans. In Python, ll the operators are special keywords or symbols tht perform operations on variables and vaues. 
The 7 types of operators are ; 
1. Arithmetic operators 
2. Assignment Opearator 
3. Comparison operator 
4. logical Operator 
5. Membership Operator 
6. Identity operator 
7. Bitwise Operators 

Arithmetic Operator : Arithmetic Operators are those which are used to perform the opertions of +,-, *, / 
+ : Addition oof 2 Operands 
- : Subtraction of 2 Operands
* : Multiplication of 2 Operands 
/ : Division of 2 Operands 
% : Modulus Operator for showing the remainder 
// : Returns the quotient, removing the decimal  part 
** : It is for performing the exponents  

comparison Operators are the operators that assign the values whether they are true or false based on the comparison. 
1. == (Equal) : retuens True if both operands are equal 
2. != (Not Equal) : Returns True if the operands are not equal 
3. > (Greater Than) : Returns true when the left operand is greater than right operand 
4. < (Less than) : Returns False when the right operand is less than left operand 
5. >= : Return true when the left operand is greater than or equal to the right operand 
6.<= : Returns true if the left operand is less than or equal to the right operand 


Assignment Operators is the operator which is used to assign values to the operator  
For example: =, +=, -=, /=, 
= is used to define the operator as a=10 
a+=b is  used to define the a= a+b   

Logical Operators are those which are used to make relation and the conditional statements. 
For example: and returns true if both the statements are true 
or returns false when both the statements are false 
not : Returns true if both the statements are false '

Bitwise Operators are those which perform operations on both  binary representation of integers 
& : Bitwise AND 
| : Bitwise OR 
^ : Bitwise XOR 
- : Bitwise NOT 
<< : Left Shift 
>> : Right Shift 

Membership Operator : Membersip Operatos are used to test if a sequence is presented in an object. 
in : Returns True if  the sequence is present in the list 
not in : Returns true if the sequence is not present in the list 

Identity Operator : I is used to compare the memory locatipn of 2 operands. 
is : Returns true if both the operands are of the same object 
is not : Returns true if both the operands are of not the same object

```python
#2 calculate Percentage 
total_marks = sum_of_marks 
percentage =  total_marks / 5 
print("Total Marks: ", total_marks)
print("Percentage", percentage)  
```

Q9 .Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system.
Given:
mass_earth = 5.972e24  # Mass of Earth in kilograms
mass_moon = 7.34767309e22  # Mass of Moon in kilograms
mass_sun = 1.989e30  # Mass of Sun in kilograms

distance_earth_sun = 1.496e11  # Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8  # Average distance between Moon and Earth in meters

Tasks:

 1. Calculate the gravitational force between the Earth and the Sun.
  2. Calculate the gravitational force between the Moon and the Earth.
 3. Compare the calculated forces to determine which gravitational force is stronger.Explain which celestial body  (Earth or Moon)    is more attracted to the other based on the comparison.



```python
# The code for the above Question is : 
mass_earth = 5.972e24  
mass_moon = 7.34767309e22  
mass_sun = 1.989e30  

distance_earth_sun = 1.496e11  
distance_moon_earth = 3.844e8  

G = 6.67430e-11 

force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun**2) 
force_moon_earth = G * (mass_earth * mass_moon) / (distance_moon_earth**2) 

print("Gravitational Force between the earth and sun: ", force_earth_sun)
print("Gravitational force between the moon and the earth: ", force_moon_earth) 

if force_earth_sun > force_moon_earth : 
    print("The force between Earth and Sun is stronger")

else: print("The gravitational force between the Moon and Earth is stronger")
```
