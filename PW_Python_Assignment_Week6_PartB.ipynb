Q1. Implement the abstract class Shape with an abstract method. 

Calculate the area. create the subclasses such as Rectangele and Circle that inherit from the Shape and override the calculate area.
Method to calculate the area of respective shapes. 

```python
from abc import ABC, abstractmethod 
import math  

# Absstract Class Shape
class Shape (ABC): 
    @abstractmethod 
    def calculate_area(self): 
        pass  

# Concrete class Rectangle 

class Rectangle(Shape):
    def __init__(self, width, length): 
        self.length =length
        self.width = width 

    def calculate_area(self): 
        return self.length * self.width 

    
# Concrete Class Circle v

class Circle(Shape): 
    def __init__(self,radius): 
        self.radius = radius 

    def calculate_area(self): 
        return math.pi * self.radius ** 2  

# Example usage 

def main(): 
    rectangle = Rectangle(15.0, 7.0) 
    circle = Circle(2.5)

    # Calculate and print the area  
    print("Rectangle Area: ", rectangle.calculate_area())
    print("Circle Area: ", circle.calculate_area())

if __name__ == "__main__":
    main()
```

Q3.  Define a class Calculator with overloaded methods add() to perform addition with different number of arguments. 
Test the calculator by adding numbers using different number of arguments 

```python
class Calculator : 
    def add (self, *args): 

        """
        Overloaded add method can handle different number of arguments
        """

        total = 0 
        for arg in args : 
            total += arg 
        return total 

def main(): 
    calculator = Calculator()  

    result1 = calculator.add(5,3)
    print("Adding 2 numbers: ", result1) 

    result2 = calculator.add(2,4,6)
    print("Adding 3 numbers; ", result2)

    result3 = calculator.add(1,2,3,4)
    print("Adding 4 numbers: ", result3)

if __name__ == "__main__":
    main()
```

Q4. Implement a Class Department with another class Employee. The department class should contain a list of Employee objects as

attribute. Demonstrate the aggregation by adding employees to different departments. 

```python
class Employee : 
    def __init__(self, name ,position): 
        self.name = name 
        self.position = position 

    def __str__(self): 
        return f"{self.name} ({self.position})"  

class Department : 
    def __init__(self,name): 
        self.name = name 
        self.employees = [] 

    def add_employee(self,employee): 
        self.employees.append(employee) 

    def __str__(self): 
        employees_str = " , ".join(str(emp) for emp in self.employees) 
        return f"{self.name} Department:\n{employees_str}"  

# Example usage 
def main(): 
    # Create some employees 
    emp1 = Employee("Master ji", "Data Scientist") 
    emp2 = Employee ("AK Jadhav", "Project Manager")
    emp3 = Employee ("T.Raza", "Senior Developer")
    emp4 = Employee("Vishwa Mohan", "CTO")
    emp5 = Employee ("Kanchan", 'HR')

    # Create some departments  
    it_department = Department("IT") 
    hr_department = Department("Human Departments") 

    # Add  employees to departments  
    it_department.add_employee(emp1)
    it_department.add_employee(emp2)
    it_department.add_employee(emp3)
    hr_department.add_employee(emp4) 
    hr_department.add_employee(emp5) 

    print(it_department)
    print("\n", hr_department)
    

if __name__ == "__main__" :
          main()


# Hnce we get the result of the code 
```

Q5. Design classes representing a House and its various components like Bedroom, LivingRoom, and Kitchen. 
       Implement composition  by   creating a House object that contains instances of its components.

```python
class Bedroom : 
    def __init__(self,name,size): 
        self.name = name 
        self.size = size 

    def __str__(self): 
        return f"{self.name} (Size : {self.size} sq ft)" 

class LivingRoom: 
    def __init__(self,size) : 
        self.size = size 

    def __str__(self): 
        return f"Living Room (Size : {self.size} sq ft)" 

class Kitchen : 
    def __init__(self, size, appliances): 
        self.size = size 
        self.appliances = appliances 

    def __str__(self): 
        appliances_str = " , ".join(self.appliances) 
        return f"Kitchen (Size: {self.size} sq ft, Appliances : {appliances_str})" 

class House : 
    def __init__(self, address, bedroom, living_room, kitchen): 
        self.address = address 
        self.bedroom = bedroom
        self.living_room = living_room 
        self.kitchen = kitchen 

    def __str__ (self): 
        bedroom_str = "\n".join(str(bedroom) for bedroom in self.bedroom)
        return f"House at {self.address} : \n{bedroom_str}\n {self.living_room}\n {self.kitchen}" 

# Example usage  

def main(): 
    bedroom1 = Bedroom("Master Bedroom", 250)
    bedroom2 = Bedroom("Guest Bedroom", 180)

    living_room = LivingRoom(400) 

    kitchen = Kitchen (150, ["Srove", "Refrigerator", "Dishwasher"]) 

    # Create the house  
    house = House("123 main Street bangalore IN", [bedroom1, bedroom2], living_room, kitchen) 

    # Print the house information 
    print(house)

if __name__ == "__main__" : 
    main()
```

Q6.Create a base class Shape with a method area() to calculate the area of different shapes. Implement subclasses like Rectangle and      
       Circle that inherit from Shape and override the area() method with appropriate formulas.


```python
import math 

# base Class Shape 
class Shape : 
    def area(self): 
        pass 

# Subclass Rectangle 
class Rectangle(Shape): 
    def __init__(self, length, width): 
        self.length = length 
        self.width = width 

    def area(self): 
        return self.length * self.width 

# Subclass Circle  
class Circle(Shape): 
    def __init__(self,radius):
        self.radius = radius 

    def area(self): 
        return math.pi * self.radius ** 2  

def main(): 
    rectangle = Rectangle(5,3)
    circle = Circle(2.5) 

    # Calculate and print the areas 
    print("Rectangle Area: ", rectangle.area()) 
    print("Circle Area: ", circle.area()) 

if __name__ == "__main__" :
    main()
```

Q7. Define a class Math with static methods for common mathematical operations like addition, subtraction, multiplication, and division. 
      Also, implement a class method to compute the factorial of a number. 

```python
class Math : 
    @staticmethod 
    def add(a,b):
        return a + b 

    @staticmethod 
    def subtract(a,b):
        return a - b

    @staticmethod 
    def multiply (a,b): 
        return a*b

    @staticmethod 
    def divide (a,b):
        if b ==0 : raise ValueError ("cannot divide by zero")
        return a/b

    @classmethod 
    def factorial(cls,n): 
        if n < 0 : 
            raise ValueError ("Factorial is not defined for negative numbers")

        if n == 0 : 
            return 1 
        return n * cls.factorial(n-1) 

    def main(): 
        print("Welcome to Maths Operation Program")

        while True : 
            print("Choose an Operation")
            print("1. Addition")
            print("2. Subtraction")
            print("3. Multiplication")
            

def main():
    print("Welcome to Math Operations Program!")

    while True:
        print("\nChoose an operation:")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Factorial")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '6':
            print("Thank you for using the program. Goodbye!")
            break

        if choice in ['1', '2', '3', '4']:
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))

                if choice == '1':
                    result = Math.add(num1, num2)
                    print("Result:", result)
                elif choice == '2':
                    result = Math.subtract(num1, num2)
                    print("Result", result) 
                elif choice == '3':
                    result = Math.multiply(num1,num2)
                    print("Result", result)
                elif choice == '4' :
                    result =  Math.divide (num1, num2)
                    print("Result", result)

            except ValueError : 
                print("Invalid error! Please enter valid numbers")

        elif choice == '5' : 
            try : 
                num = int(input("Enter thenumber to calculate the factorial"))
                result = Math.factorial(num)
                print("Factorial: ", result)

            except ValueError:
                print("Invalid Input ! Please enter a valid integer")

        else : 
            print("Invalid Choice. Pleae enter a number number between 1 and 6")
    
if __name__ == "__main__" : 
    main()

    

```
